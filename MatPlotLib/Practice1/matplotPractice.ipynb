```python
batter =pd.read_csv('batter.csv')
```

```python
batter.head()
```

```python
sample_df=batter.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
#labeling 

x=[1,2,3,4]
y=[3,5,6,7]
plt.scatter(x,y)
plt.text(1,3,'point1')
plt.text(2,5,'point2')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
#with runs
plt.figure(figsize=(10,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
#horizontal line by setting y value

#with runs
plt.figure(figsize=(10,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],sample_df['runs'])
plt.axhline(120,color='red')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
#horizontal,vertical line  line by setting y value,x value

#with runs
plt.figure(figsize=(10,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],sample_df['runs'])
plt.axhline(120,color='red')
plt.axvline(25,color='green')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```
